```python
import matplotlib.pyplot as plt
import numpy as np
import random


#Making the figure a certain size
f = plt.figure()
f.set_figwidth(6.4*3)
f.set_figheight(4.8*3)


#setting up the triangle
tx = np.array([0, 100, 100, 0,0])
ty = np.array([0, 0, 100,100, 0])


#Getting like a certain point to be the starting point
# I kind did the starting point by hand cuz I was laazzzzyy
start = np.array([50,50])
plt.scatter(*start)
last = -1

#The actual calculation!
# The math!
# So you get ur starting point and a random vertex
# Then you find the midpoint of those two and plot that point 
# Then that plotted point is the new starting point
for i in range(10000):
    random_num = random.randint(0,3)
    if random_num != last:
        point = np.array([(start[0]+tx[random_num])/2, (start[1]+ty[random_num])/2])
        plt.scatter(*point)
    start = point
    last = random_num

#Plotting jazz
plt.title("Square!!")
plt.plot(tx,ty)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

#Generalized, kinda dog but eh
#TBH i dont think this math is right for other shapes but oh welllll

def serpenski_fractal(iter, shapex, shapey, shapeStart, nameShape="quirky"):
    f = plt.figure()
    f.set_figwidth(6.4*3)
    f.set_figheight(4.8*3)
    plt.plot(shapex,shapey)
    start = shapeStart
    plt.scatter(*start)
    last = -1

    for i in range(iter):
        random_num = random.randint(0,len(shapex)-1)
        if (len(shapex)-1) == 3:
            point = np.array([(start[0]+shapex[random_num])/2, (start[1]+shapey[random_num])/2])
            plt.scatter(*point)
        elif random_num != last:
            point = np.array([(start[0]+shapex[random_num])/2, (start[1]+shapey[random_num])/2])
            plt.scatter(*point)
        start = point
        last = random_num
    
    plt.title(f'{nameShape} Fractal!!')
    plt.show()


tx = np.array([0, 100, 50, 0])
h = np.sqrt(3)*100/2
ty = np.array([0, 0, h,0])
start = np.array([tx[2], ty[2]/2])
serpenski_fractal(100, tx, ty, start, "Triangle")

tx = np.array([0, 100, 100, 0,0])
ty = np.array([0, 0, 100,100, 0])
start = np.array([50,50])
serpenski_fractal(1000, tx, ty, start, "Square")

# tx = np.array([-12, -6, 6, 12, 6, -6, -12]) *10
# ty = np.array([0, -6, -6, 0, 6, 6, 0]) * 10
# start = np.array([2,4]) * 10
# serpenski_fractal(1000, tx, ty, start, "Hex")
```
